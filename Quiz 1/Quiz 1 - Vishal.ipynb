## Task 2

```python
date = df.SampleDate.values.tolist()
date = list(set(date))
parameter = df.Parameter.values.tolist()
characteristic = df.Characteristic.values.tolist()
```

```python
combined = [i+" "+j for i,j in zip(parameter, characteristic)]
df['Parameter'] = combined
```

```python
combined = list(set(combined))
```

```python
df = df.drop(columns=['WBodyID', 'WaterBodyName', 'DataSource', 'StationID', 'StationName',
       'Actual_StationID', 'Actual_Latitude', 'Actual_Longitude', 'DEP_WBID',
        'ActivityDepth', 'DepthUnits', 
       'Characteristic', 'Sample_Fraction',
       'QACode', 'Result_Comment', 'Original_Result_Value',
       'Original_Result_Unit'])
```

```python
df.head(5)
```

```python
#creating a dataframe of relevant data for the task
reorg =list()
for d in date:
    top = {key:() for key in combined}
    for i,j in df.iterrows():
        if(j[0]==d):
            top[j[1]]=(j[2],j[3])
    reorg.append(top)
```

```python
new_df = pd.DataFrame(data=None, columns=["Timestamp"])
```

```python
new_df['Timestamp'] = date
```

```python
for i in combined:
    top = list()
    for j in reorg:
        top.append(j[i])
    new_df[i] = top
```

```python
new_df.head(5)
```

```python
#Except timestamp, all other features are different parameters
new_df.columns
```

```python
#Creating a new dataframe with normalised values
newdf_values = new_df.copy()
for i in combined:
    unit = list()
    for j in newdf_values[i]:
        try:
            unit.append(j[1])
        except:
            unit.append("")
    unit_match = list(set(unit))
    if('ug/l' in unit_match and 'mg/l' in unit_match):
        vals = list()
        for j in newdf_values[i]:
            try:
                if(j[1].count("mg/l")>0):
                    vals.append(j[0]*1000)
                else:
                    vals.append(j[0])
            except:
                vals.append("NaN")
    elif("mg/l" in unit_match):
        vals = list()
        for j in newdf_values[i]:
            try:
                vals.append(j[0]*1000)
            except:
                vals.append("NaN") 
    else:
        vals = list()
        for j in newdf_values[i]:
            try:
                vals.append(j[0])
            except:
                vals.append("NaN")
    newdf_values[i] = vals  
```

```python
#Sorting the date
temp = newdf_values.Timestamp.values.tolist()
temp = [i.split(" ")[0].replace("-","/") for i in temp]
format_str = '%m/%d/%Y'
temp = [datetime.datetime.strptime(str(i), format_str) for i in temp]
newdf_values['Date'] = temp
newdf_values = newdf_values.sort_values(by=['Date'])
```

```python
tempdf = newdf_values[newdf_values['NH3_N_diss_ugl Nitrogen, ammonia as N']!="NaN"]
plt.figure(figsize=(10,5))
plt.plot(tempdf['Timestamp'], tempdf['NH3_N_diss_ugl Nitrogen, ammonia as N'])
plt.xticks(rotation = 90)
plt.xlabel("Timestamp")
plt.ylabel("NH3_N_diss_ugl Nitrogen, ammonia as N")
plt.show()
```
